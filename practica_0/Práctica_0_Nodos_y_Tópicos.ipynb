### 1. La estructura de paquetes en ROS 2
En ROS 2, un paquete es la unidad básica de organización del código. La estructura típica de un paquete en Python incluye:
*   `package.xml`: Metadatos del paquete (nombre, versión, dependencias).
*   `setup.py`: Instrucciones de instalación.
*   `setup.cfg`: Configuración para herramientas de construcción.
*   `src/`: Directorio raíz de código fuente (opcional en el workspace).
*   Carpeta con el nombre del paquete: Contiene el código fuente (`__init__.py`, archivos `.py`).

### 2. Definición de Nodos y Tópicos
*   **Nodo**: Es un proceso ejecutable que realiza computación. Un sistema robótico suele estar compuesto por muchos nodos que se comunican entre sí. La biblioteca `rclpy` permite crear estos nodos en Python heredando de la clase `Node`.
*   **Tópico (Topic)**: Es un bus de comunicación por el cual los nodos intercambian mensajes mediante un modelo de **publicación/suscripción**. Los tópicos son anónimos; el publicador no sabe quién escucha y el suscriptor no sabe quién envía.

### 3. Conceptos Técnicos de rclpy
*   **Temporizadores (Timers)**: Se utilizan para ejecutar una función de *callback* a una frecuencia determinada. En el publicador, permite emitir mensajes periódicamente.
*   **QoS (Quality of Service)**: Define cómo se manejan los mensajes en la red. El parámetro entero al crear un publicador o suscriptor representa la profundidad de la cola (queue depth).

## Desarrollo de la Práctica

En esta práctica se desarrollaron dos nodos principales en Python utilizando la biblioteca `rclpy`.

### 1. Nodo Publicador (`primer_nodo.py`)
Este nodo genera datos aleatorios con `random.uniform(0.0, 1.0)` y los envía a un tópico llamado `topic_1` cada segundo.

````python
# Fragmento de código del Nodo Publicador
import rclpy
from rclpy.node import Node
from std_msgs.msg import Float64
import random

class MyNode(Node):
    def __init__(self):
        super().__init__("publisher_node")
        # El '10' es la profundidad de la cola de QoS
        self.publisher_ = self.create_publisher(Float64, 'topic_1', 10)
        # Temporizador a 1Hz (1.0 segundos)
        self.timer = self.create_timer(1.0, self.cbck)

    def cbck(self):
        msg = Float64()
        msg.data = random.uniform(0.0, 1.0)
        self.publisher_.publish(msg)
        self.get_logger().info(f"Publicando en topic_1: {msg.data:.3f}")
````

### 2. Nodo Suscriptor (`subscriber_node.py`)
Este nodo recibe la información de `topic_1` y la procesa para mostrar el valor transformado a una magnitud física (ejemplo: radianes).

````python
# Fragmento de código del Nodo Suscriptor
import math
from rclpy.node import Node
from std_msgs.msg import Float64

class SubscriberNode(Node):
    def __init__(self):
        super().__init__("subscriber_node")
        # Se suscribe al mismo tópico y tipo de mensaje
        self.create_subscription(Float64, 'topic_1', self.sub_cbck, 10)

    def sub_cbck(self, msg):
        # Transformación solicitada: Conversión a rango [0, 2π]
        rad = msg.data * 2 * math.pi
        self.get_logger().info(f"Recibido: {msg.data:.3f} -> Transformado: {rad:.3f} rad/s")
````

## Resultados

Se programó exitosamente un Nodo Publicador que envía mensajes `Float64` en un rango de [0, 1]. El Nodo Suscriptor recibe estos mensajes y realiza la transformación matemática correspondiente.
